```python
#Definimos la paleta de colores de Matpltlib
plt.style.use('seaborn-muted')
```

```python
#plt.style.use('seaborn-pastel')
sns.countplot(data= df, x= 'Genero')
plt.title('Muestra de Genero')
```

```python
sns.barplot (x = 'Genero', y = 'Edad', data = df, estimator= np.median)
plt.title('Estimador Media Genero')
```

```python
sns.barplot (x = 'Genero', y = 'Edad', data = df, estimator= np.max)
plt.title('Estimador Maximo Genero')
```

```python
sns.barplot (x = 'Genero', y = 'Edad', data = df, estimator= np.min)
plt.title('Estimador Minimo Genero')
```

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10), sharex=False)

sns.countplot (x = 'Genero', data = df, ax = ax1)
sns.barplot (x = 'Genero', y = 'Edad', data = df, estimator= np.median, ax= ax2)
sns.barplot (x = 'Genero', y = 'Edad', data = df, estimator= np.max, ax= ax3)
sns.barplot (x = 'Genero', y = 'Edad', data = df, estimator= np.min, ax=ax4)

ax1.set_title('Muestra de Genero')
ax2.set_title('Estimador Media Genero')
ax3.set_title('Estimador Máximo Genero')
ax4.set_title('Estimador Mínimo Genero')

plt.savefig('Estimadores1.jpg')
```

```python
plt.style.use('ggplot')
sns.set_theme(style="white", context="poster")
sns.distplot(df['Edad'], bins = 25)
```

```python
edu = df.groupby(df['Nivel educativo'])['Genero'].count().sort_values(ascending=False)
```

```python
edu
```

```python
sns.set_theme(style = "ticks", context = "paper")
plt.bar(edu.index,  edu.values)
plt.xticks(rotation=25, ha='right')
plt.show()
```

```python
print(plt.style.available)
```

```python
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))

patches, texts, autotexts = plt.pie(edu.values, labels= edu.index,
       startangle=180,
       autopct= '%.1f%%',
        rotatelabels = True,
       labeldistance = None,
       textprops={'size': 'small'},)

#Etiquetas (:
ax.legend(title="Nivel máximo de estudios",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))


#Presentacion del Plot
plt.axis('equal')
plt.title('Nivel máximo de estudios', color='k')
plt.savefig('Educacion.png', bbox_inches='tight')
#plt.imsave('new_1.png', arr='dimensional')
```
