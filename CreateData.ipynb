```python
sabesnutricion = []
for x in itertools.repeat('No conozco nada en nutrición', 14):
  sabesnutricion.append(x)
for x in itertools.repeat('Mi nivel de conocimientos no es muy bueno', 289):
  sabesnutricion.append(x)
for x in itertools.repeat('No conozco nada en nutrición', 554):
  sabesnutricion.append(x)
for x in itertools.repeat('No conozco nada en nutrición', 144):
  sabesnutricion.append(x)
```

```python
def mezclar_lista(lista_original):
    # Crear una copia, ya que no deberíamos modificar la original
    # https://parzibyte.me/blog/2020/05/31/python-clonar-lista-eliminar-referencia/
    lista = lista_original[:]
    # Ciclo for desde 0 hasta la longitud de la lista -1
    longitud_lista = len(lista)
    for i in range(longitud_lista):
        # Obtener un índice aleatorio
        # https://parzibyte.me/blog/2019/04/04/generar-numero-aleatorio-rango-python/
        indice_aleatorio = random.randint(0, longitud_lista - 1)
        # Intercambiar
        temporal = lista[i]
        lista[i] = lista[indice_aleatorio]
        lista[indice_aleatorio] = temporal
    # Regresarla
    return lista
```

```python
Genero = mezclar_lista(genero);
Edad = mezclar_lista(edad);
Educacion = mezclar_lista(educacion);
Ingresos = mezclar_lista(ingresos);
Responsable = mezclar_lista(responsable);
Nutricion = mezclar_lista(nutricion);
Sabes = mezclar_lista(sabesnutricion)
```

```python
df_listas = []
df_listas.append(Genero)
df_listas.append(Edad)
df_listas.append(Educacion)
df_listas.append(Ingresos)
df_listas.append(Responsable)
df_listas.append(Nutricion)
df_listas.append(Sabes)
print(df_listas)
```

```python
df = pd.DataFrame(df_listas)
```
